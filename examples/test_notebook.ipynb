# Example Notebook
This example notebook shows how to use the caispp package.  Specifically, the notebook trains an image classifier and saves it.  It then loads the classifier and uses it to predict the label of a new image.

```python
from caispp import ImageDataset, ImageClassifier, Path
```

```python
path = Path('example_dataset/')
dataset = ImageDataset(path, show_distribution=True)
```

```python
classifier = ImageClassifier(dataset)
```

```python
classifier.train(epochs=10, show_output=False)
```

```python
classifier.show_history()
```

```python
classifier.test(show_distribution=True)
```

```python
classifier.save('classifier')
```

```python
model = ImageClassifier()
model.load('classifier')
```

```python
preds = model.predict_from_paths(['images/picture.jpg'])
```

```python
for pred in preds:
    print('Predicted', pred[0], 'with confidence' ,pred[1])
```
